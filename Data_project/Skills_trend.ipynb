```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

# Filtering for Data Analyst job in the USA

```python
df_da_us = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
df_da_us['job_posted_month_no'] = df_da_us['job_posted_date'].dt.month

# Explode the job skills column
df_da_us_explode = df_da_us.explode('job_skills')

# Creating a pivot table
df_da_pivot = df_da_us_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_da_pivot
```

```python
# Create a new column for Total
df_da_pivot.loc['Total'] = df_da_pivot.sum()
df_da_pivot = df_da_pivot[df_da_pivot.loc['Total'].sort_values(ascending=False).index]
df_da_pivot = df_da_pivot.drop('Total')
df_da_pivot
```

```python
# Plotting
da_totals = df_da_us.groupby('job_posted_month_no').size()
```

```python
df_da_us_percent = df_da_pivot.div(da_totals/100, axis=0)
df_da_us_percent
```

```python
# Changing month number to month name
df_da_us_percent = df_da_us_percent.reset_index()
df_da_us_percent['job_posted_month'] = df_da_us_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_da_us_percent = df_da_us_percent.set_index('job_posted_month')
df_da_us_percent = df_da_us_percent.drop(columns='job_posted_month_no')
df_da_us_percent
```

```python
# Plotting
df_plot = df_da_us_percent.iloc[:, :5]

# Creating Seaborn plot
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills of Data Analyst in the USA')
plt.ylabel('Likelhood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
