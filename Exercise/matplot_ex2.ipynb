```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Plotting work from home vs. not work from home
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Work from Home Status')
plt.ylabel('')
plt.show()
```

```python
# Plotting for Job Titles
df['job_title_short'].value_counts().plot(kind='pie')
```

```python
# Comparing three different values
df[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
```

```python
# Creating subplots for above values using matplot subplots
fig, ax = plt.subplots(1, 3)

# Creating Dictionary list for values
dict_colum = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance'
}

for i, (column, title) in enumerate(dict_colum.items()):
    ax[i].pie(df[column]).value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangel=90
    ax[i].set_title(title)

plt.show()
```
