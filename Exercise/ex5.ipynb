```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Creating month
df['job_poted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
# Create a list of different months
months = df['job_poted_month'].unique()
dict_months = {month: df[df['job_poted_month'] == month] for month in months}
dict_months['Jan']
```

```python
# Concatenate the dataframes for first quarter
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
# Value counts of first quarter
df_q1['job_poted_month'].value_counts().plot(kind='bar')
```

```python
# Export Data to Excel
df_q1.to_excel('ex5.xlsx')
```
