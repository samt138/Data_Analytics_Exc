```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Get the data analyst job in india
df_da_ind = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'India')].copy()

# Drop Not A Number Values from the dataset
df_da_ind = df_da_ind.dropna(subset=['salary_year_avg'])

df_da_ind
```

```python
# Plotting the box
df_da_ind['salary_year_avg'].plot(kind='box', vert=False)
plt.title('Salary Distribution for Data Analyst Jobs in India')
plt.ylabel('Salary')
plt.show()
```

```python
# Comparing jobs in data field
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

# Filter the data
df_ind = df[(df['job_title_short'].isin(job_titles)) & (df['job_country'] == 'India')].copy()
df_ind = df_ind.dropna(subset=['salary_year_avg'])

# Creating a for loop for Average yearly salary
job_list = [df_ind[df_ind['job_title_short'] == job_titles]['salary_year_avg'] for job_titles in job_titles]
```

```python
# Creating boxplot
plt.boxplot(job_list, labels=job_titles, vert=False)
plt.title('Salary Distribution in India')
plt.xlabel('Year Salary INR')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'{int(y/1000)}K'))
plt.xlim(0, 600000)
plt.show()
```
