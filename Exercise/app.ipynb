```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_poseted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Getting top 10 Results
df.iloc[:10]
```

```python
# Median Salary
median_salary_year = df['salary_year_avg'].median()
```

```python
# By Hours
median_salary_hour = df['salary_hour_avg'].median()
```

```python
# Filter Data
filt = df
filt['salary_year_avg'] = filt['salary_year_avg'].fillna(median_salary_year)
filt['salary_hour_avg'] = filt['salary_hour_avg'].fillna(median_salary_hour)
```

```python
# Top 10 Results
filt.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
# Dropping Duplicates
filt_unique = filt
filt_unique = filt.drop_duplicates()
```

```python
# Removing Duplicate Job Titles
filt_unique = filt_unique.drop_duplicates(subset=['job_title', 'company_name'])

print('Original df: ', len(filt))
print('Duplicates dropped: ', len(filt_unique))
print('Rows Dropped: ', len(filt) - len(filt_unique))
```

```python
# Copy
df_copy = df.copy()

print('ID of Original df: ', id(df))
print('ID of Copy df: ', id(df_copy))
print('Are these the same: ', id(df) == id(df_copy))
```
